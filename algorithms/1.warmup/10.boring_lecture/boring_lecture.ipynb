```python
filename = "boring_lecture_input_2.txt"

with open(filename, encoding="utf-8") as src:
    string = src.readline().strip()

answer = {}
for letter_idx in range(len(string)):
    answer[string[letter_idx]] = answer.get(string[letter_idx], 0) + \
          (len(string[:letter_idx]) + 1) * len(string[letter_idx:])

for letter in sorted(answer):
    print(letter, ": ", answer[letter], sep='')
```
