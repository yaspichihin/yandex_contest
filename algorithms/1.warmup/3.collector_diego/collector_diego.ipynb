```python
with open("collector_diego_input_2.txt", encoding="utf-8") as src:
    n = int(src.readline())
    stickers = sorted(set(map(int, src.readline().split())))
    k = int(src.readline())
    min_numbers = list(map(int, src.readline().split()))


for min_number in min_numbers:
    l = 0
    r = len(stickers) - 1
    while l < r:
        m = (l + r + 1) // 2
        if stickers[m] < min_number:
            l = m
        else:
            r = m - 1

    if stickers[l] < min_number:
        l += 1
    
    print(l)
```
