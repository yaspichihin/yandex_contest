```python
filename = "sum_in_rectangle_input_2.txt"

with open(filename, encoding="utf-8") as src:
    n, m, k = map(int, src.readline().strip().split())

    prefix_sum_2d = [[0] * (m + 2) for _ in range(n + 2)]
    for i in range(n):
        string = src.readline().strip().split()
        for j in range(m):
            prefix_sum_2d[i+1][j+1] = int(string[j]) + \
                                        prefix_sum_2d[i+1][j] + \
                                        prefix_sum_2d[i][j+1] - \
                                        prefix_sum_2d[i][j]

    points = [list(map(int, src.readline().strip().split())) for _ in range(k)]

answer = []
for point in points:
    x1, y1, x2, y2 = point
    answer.append(prefix_sum_2d[x2][y2] - prefix_sum_2d[x2][y1-1] - prefix_sum_2d[x1-1][y2] + prefix_sum_2d[x1-1][y1-1])

with open("output.txt", "w", encoding="utf-8") as dst:
    for line in answer:
        dst.write(str(line) + "\n")
```
