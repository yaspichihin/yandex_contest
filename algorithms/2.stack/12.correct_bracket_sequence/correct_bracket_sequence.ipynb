```python
class Node:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

class Stack:
    def __init__(self):
        self.last = None
        self.length = 0

    def push(self, value: int) -> None:
        new_node = Node(value)
        if self.length == 0:
            self.last = new_node
            self.length += 1
        else:
            self.last.next = new_node
            new_node.prev = self.last
            self.last = new_node
            self.length += 1

    def check_stack(self) -> bool:
        return True if self.length else False

    def pop(self) -> str:
        if self.check_stack():
            value = self.last.value
            if self.length == 1:
                self.last = None
            else:
                new_last = self.last.prev
                self.last.prev = None
                new_last.next = None
                self.last = new_last
            self.length -= 1
            return value

    def back(self) -> str:
        if self.check_stack():
            return self.last.value

    def size(self) -> int:
        return self.length


stack = Stack()

with open("./correct_bracket_sequence_input_3.txt", encoding="utf-8") as src:
    seq = src.readline().strip()

answer = "yes"

brackets_dict = {")": "(",
                 "]": "[",
                 "}": "{"}

for bracket in seq:
    if bracket in "([{":
        stack.push(bracket)
    elif stack.back() == brackets_dict[bracket]:
        stack.pop()
    else:
        answer = "no"
        break
if stack.size() != 0:
    answer = "no"

print(answer)
```
