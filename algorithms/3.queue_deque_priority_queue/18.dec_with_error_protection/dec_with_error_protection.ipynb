```python
class Node:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

class Deque:
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0

    def push_front(self, value) -> None:
        node = Node(value)
        if self.length == 0:
            self.tail = node
        else:
            node.next = self.head
            self.head.prev = node
        self.head = node
        self.length += 1
        print("ok")
        
    def push_back(self, value) -> None:
        node = Node(value)
        if self.length == 0:
            self. head = node
        else:
            node.prev = self.tail
            self.tail.next = node
        self.tail = node
        self.length += 1
        print("ok")

    def check_deque(self) -> None:
        return True if self.length else False

    def pop_front(self) -> None:
        if self.check_deque():
            node = self.head
            if self.length == 1:
                self.head = None
                self.tail = None
            else:
                self.head = self.head.next
                self.head.prev = None
                node.next = None
            self.length -= 1
            print(node.value)
        else:
            print("error")

    def pop_back(self) -> None:
        if self.check_deque():
            node = self.tail
            if self.length == 1:
                self.head = None
                self.tail = None
            else:
                self.tail = self.tail.prev
                self.tail.next = None
                node.prev = None
            self.length -= 1
            print(node.value)
        else:
            print("error")

    def front(self):
        if self.check_deque():
            print(self.head.value)
        else:
            print("error")

    def back(self):
        if self.check_deque():
            print(self.tail.value)
        else:
            print("error")

    def size(self) -> None:
        print(self.length)

    def clear(self) -> None:
        self.head = None
        self.tail = None
        self.length = 0
        print("ok")

    def exit(self) -> None:
        print("bye")


deque = Deque()
with open("dec_with_error_protection_input_3.txt", encoding="utf-8") as src:
    for line in src:
        line = line.strip().split()
        
        command = line[0]
        if len(line) > 1:
            value = int(line[1])
        
        if command == "push_front":
            deque.push_front(value)
        elif command == "push_back":
            deque.push_back(value)
        elif command == "pop_front":
            deque.pop_front()
        elif command == "pop_back":
            deque.pop_back()
        elif command == "front":
            deque.front()
        elif command == "back":
            deque.back()
        elif command == "size":
            deque.size()
        elif command == "clear":
            deque.clear()
        elif command == "exit":
            deque.exit()
            del deque
            break
```
