```python
class Heap:
    def __init__(self):
        self.length = 0
        self.array = []

    def insert(self, value):
        """ Добавить число
        """
        self.array.append(value)
        self.length += 1
        position = self.length - 1
        # Просеивание вверх
        while position > 0 and \
            self.array[position] > self.array[(position - 1) // 2]:
            self.array[position], self.array[(position - 1) // 2] = \
            self.array[(position - 1) // 2], self.array[position]
            position = (position - 1) // 2


    def extract(self):
        """  Извлечь наибольшое число
        """
        maximum = self.array[0]
        self.array[0] = self.array[-1]
        position = 0
        # Просеивание вниз
        while position * 2 + 1 < self.length - 1:
            son_1_idx = position * 2 + 1
            son_2_idx = position * 2 + 2
            if self.array[son_1_idx] > self.array[son_2_idx]:
                max_son_idx = son_1_idx
            else:
                max_son_idx = son_2_idx
            if self.array[position] < self.array[max_son_idx]:
                self.array[position], self.array[max_son_idx] = \
                self.array[max_son_idx], self.array[position]
                position = max_son_idx
            else:
                break
        self.array.pop()
        self.length -= 1
        return maximum


with open("./heapuy_input_2.txt", encoding="utf-8") as src:
    n_commands = int(src.readline().strip())
    commands = [list(map(int, command.strip().split())) for command in src.readlines()]

heap = Heap()
for command in commands:
    if command[0] == 0:
        heap.insert(command[1])
    elif command[0] == 1:
        print(heap.extract())
```
