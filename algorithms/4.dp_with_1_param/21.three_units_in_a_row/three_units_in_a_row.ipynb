```python
filename = "three_units_in_a_row_input_1.txt"

with open(filename, encoding="utf-8") as src:
    n = int(src.readline())

num = 0
base = [0, 1, 1]
for i in range(3, n+3):
    # Можно было и через append,
    # но пожертвововал скоростью ради экономии памяти.
    num = base[0] + base[1] + base[2]
    base[0], base[1], base[2] = base[1], base[2], num

print(num)
```
