```python
filename = "calculator_input_3.txt"

with open(filename, encoding="utf-8") as src:
    num = int(src.readline())

prev = [0 for _ in range(num+1)]
dp = [0 for _ in range(num+1)]
dp[0] = -1

for i in range(1, num+1):
    var1 = dp[i-1] + 1
    var2 = dp[int(i/2)] + 1 if i % 2 == 0 else float("inf")
    var3 = dp[int(i/3)] + 1 if i % 3 == 0 else float("inf")


    dp[i] = min(var1, var2, var3)
    if dp[i] == var1:
        prev[i] = i-1
    elif dp[i] == var2:
        prev[i] = int(i/2)
    elif dp[i] == var3:
        prev[i] = int(i/3)

print(dp[-1])
path = [num]
while prev[num] > 0:
    path.append(prev[num])
    num = prev[num]
print(*path[::-1])
```
