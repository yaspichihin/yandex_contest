```python
filename = "buying_tickets_input_1.txt"

with open(filename, encoding="utf-8") as src:
    n = int(src.readline())    # 1 <= n <= 5_000
    abc = [list(map(int, line.split())) for line in src.readlines()]

# Добавляем бесконечность для базы
abc = [[float("inf")] * 3 for _ in range(3)] + abc
dp = [0] * 3

for i in range(3, n + 3):
    value = min(dp[i - 1] + abc[i][0],
                dp[i - 2] + abc[i - 1][1],
                dp[i - 3] + abc[i - 2][2])
    dp.append(value)
    
print(value)
```
