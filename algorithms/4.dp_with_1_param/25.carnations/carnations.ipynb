```python
filename = "carnations_input_2.txt"

with open(filename, encoding="utf-8") as src:
    n = int(src.readline())
    coords = list(map(int, src.readline().split()))

coords_sorted = sorted(coords)

# Добавим базу
dp = [0 for _ in range(n)]

for i in range(n):
    if i == 0:
        dp[i] = 0
    elif i == 1:
        dp[i] = coords_sorted[i] - coords_sorted[i-1]
    elif i == 2:
        dp[i] = dp[i-1]  + coords_sorted[i] - coords_sorted[i-1]
    else:
        dp[i] = min(dp[i-2], dp[i-1]) + coords_sorted[i] - coords_sorted[i-1]

print(dp[-1])
```
