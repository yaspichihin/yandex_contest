```python
filename = "knight_move_input_3.txt"

with open(filename, encoding="utf-8") as src:
    n, m = map(int, src.readline().split())

pole = [[-2  for _ in range(m+2)] for _ in range(n+2)]
pole[0][1] = 0
pole[1][0] = 0
pole[2][2] = -2

for i in range(2, n+2):
    for j in range(2, m+2):
        if pole[i-1][j-2] != -2 and pole[i-2][j-1] != -2:
            pole[i][j] = pole[i-1][j-2] + pole[i-2][j-1]

        elif pole[i-1][j-2] != -2:
            pole[i][j] = pole[i-1][j-2] + 1 if i == 3 and j == 4 else pole[i-1][j-2]

        elif pole[i-2][j-1] != -2:
            pole[i][j] = pole[i-2][j-1] + 1 if i == 4 and j == 3 else pole[i-2][j-1]


# for i in range(2, n+2):
#     for j in range(2, m+2):
#         print(pole[i][j], end=" ")
#     print()

print(1 if n == 1 and m == 1 else 
      0 if pole[-1][-1] == -2 else 
      pole[-1][-1])
```
