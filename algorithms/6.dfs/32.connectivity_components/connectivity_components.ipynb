```python
import sys
sys.setrecursionlimit(200000)

filename = "connectivity_components_input_16.txt"

with open(filename, encoding="utf-8") as src:
    n, m = map(int, src.readline().split())
    edges = [list(map(int, edge.split())) for edge in src.readlines()]

connectivity_list = {}
visited_vertexes = {}
for i in range(1, n+1):
    connectivity_list[i] = []
    visited_vertexes[i] = False

for edge in edges:
    vertex_1, vertex_2 = edge
    connectivity_list[vertex_1] += [vertex_2]
    if vertex_1 != vertex_2:
        connectivity_list[vertex_2] += [vertex_1]

def dfs(graph, visited, now, component):
    visited[now] = (component, True)
    for neighbor in graph[now]:
        if not visited[neighbor]:
            dfs(graph, visited, neighbor, component)

component = 0
for vertex, is_visited in visited_vertexes.items():
    if not is_visited:
        component += 1
        dfs(connectivity_list, visited_vertexes, vertex, component)

answer = {}
for vertex, data in visited_vertexes.items():
    component, is_visited = data
    answer[component] = answer.get(component, []) + [vertex]

print(len(answer))
for component, data in answer.items():
    print(len(data))
    print(*data)
```
