```python
filename = "cheating_input_9.txt"

with open(filename, encoding="utf-8") as src:
    n_v, n_e = map(int, src.readline().split())
    edges = [list(map(int, row.split())) for row in src.readlines()]

connectivity_list = {}
visited_vertexes = {}
for vertex in range(1, n_v + 1):
    connectivity_list[vertex] = []
    visited_vertexes[vertex] = False

for edge in edges:
    vertex_1, vertex_2 = edge
    connectivity_list[vertex_1] += [vertex_2]
    if vertex_1 != vertex_2:
        connectivity_list[vertex_2] += [vertex_1]

flag = False
def dfs(graph, visited, now, group):
    visited[now] = (group, True)
    for neigbor in graph[now]:
        if visited[neigbor] and visited[neigbor][0] == group:
            global flag
            flag = True

        if not visited[neigbor]:
            group = 3 - group
            dfs(graph, visited, neigbor, group)


for vertex, is_visited in visited_vertexes.items():
    if not is_visited:
        group = 1
        dfs(connectivity_list, visited_vertexes, vertex, group)

print("NO" if flag else "YES")

visited_vertexes
```
