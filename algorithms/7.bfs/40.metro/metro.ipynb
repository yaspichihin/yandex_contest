```python
filename = "metro_input_1.txt"

with open(filename, encoding="utf-8") as src:
    n = int(src.readline())
    m = int(src.readline())
    lines = [src.readline().split() for _ in range(m)]
    a, b = src.readline().split()

    a = '1'
    b = '5'

conn_lst = {}
for i in range(1, m):
    conn_lst[i-1] = set(lines[i-1]) & set(lines[i])
    conn_lst[i] = set(lines[i-1]) & set(lines[i])


# Определить на каких линии находится станция a и b
flag_zero = False
for line in lines:
    if a in line and b in line:
        flag_zero = True
        break
    

print(0 if flag_zero else 1)
```
