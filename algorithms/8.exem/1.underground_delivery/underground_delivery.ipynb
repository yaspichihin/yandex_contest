```python
filename = "underground_delivery_input_2.txt"



train = []
with open(filename, encoding="utf-8") as src:
    n = int(src.readline())
    while n > 0:
        op, *other = src.readline().split()
        if op == "add":
            count = int(other[0])
            value = other[1]
            train += [[count, value]]

        elif op == "delete":
            count = int(other[0])

            while count > 0 and train:
                last = train.pop()
                if last[0] > count:
                    train += [[last[0] - count, last[1]]]
                count -= last[0]

        elif op == "get":
            value = other[0]
            answer = 0
            for wagon in train:
                if wagon[1] == value:
                    answer += wagon[0]
            print(answer)

        n -= 1
```
