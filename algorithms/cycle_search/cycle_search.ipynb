```python
filename = "cycle_search_input_1.txt"

with open(filename, encoding="utf-8") as src:
    vs = int(src.readline())
    conn_mtx = [list(map(int, row.split())) for row in src.readlines()]


conn_lst = {}
visited = {}
for v in range(1, vs+1):
    conn_lst[v] = []
    visited[v] = False

for i in range(len(conn_mtx)):
    for j in range(i+1, len(conn_mtx)):
        conn_lst[i + 1] += [j + 1]
        conn_lst[j + 1] += [i + 1]

visited
```
