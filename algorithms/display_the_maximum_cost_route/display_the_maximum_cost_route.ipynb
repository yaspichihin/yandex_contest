```python
filename = "display_the_maximum_cost_route_input_1.txt"

with open(filename, encoding="utf-8") as src:
    n, m = map(int, src.readline().split())
    mtx = [list(map(int, line.split())) for line in src.readlines()]

max_path = [[0 for _ in range(m+1)] for _ in range(n+1)]

for i in range(1, n+1):
    for j in range(1, m+1):
        max_path[i][j] = max(max_path[i-1][j], max_path[i][j-1]) + mtx[i-1][j-1]

print(max_path[i][j])

path = []
while i > 1 or j > 1:
    # print(max_path[i][j])
    if max_path[i-1][j] >= max_path[i][j-1]:
        path.append("D")
        i -= 1
    elif max_path[i-1][j] < max_path[i][j-1]:
        path.append("R")
        j -= 1
print(*path[::-1])
```
