```python
class Node:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

class Queue:
    def __init__(self):
        self.length = 0
        self.head = None
        self.tail = None
    
    def push(self, value: int) -> None:
        if self.length == 0:
            node = Node(value)
            self.head = node
            self.tail = node
            self.length += 1
        else:
            node = Node(value)
            self.tail.next = node
            node.prev = self.tail
            self.tail = node
            self.length += 1

    def check_queue(self) -> bool:
        return True if self.length else False

    def pop(self) -> int:
        if self.check_queue():
            node = self.head
            if self.length == 1:
                self.head = self.tail = None
            else:
                self.head.next.prev = None
                self.head = self.head.next
                node.next = None
            value = node.value
            self.length -= 1
            return value

    def size(self) -> int:
        return self.length

queue = Queue()

with open("drunk_game_input_1.txt", encoding="utf-8") as src:
    player_1 = Queue()
    player_2 = Queue()
    [player_1.push(num) for num in map(int, src.readline().strip().split())]
    [player_2.push(num) for num in map(int, src.readline().strip().split())]
    

count = 0

while all([player_1.size(), player_2.size()]) and count <= 1e6:
    player_1_card = player_1.pop()
    player_2_card = player_2.pop()

    if player_1_card == 0 and player_2_card == 9:
        player_1.push(player_1_card)
        player_1.push(player_2_card)
        player = "first"
    elif player_2_card == 0 and player_1_card == 9:
        player_2.push(player_1_card)
        player_2.push(player_2_card)
        player = "second"
    elif player_1_card > player_2_card:
        player_1.push(player_1_card)
        player_1.push(player_2_card)
        player = "first"
    elif player_2_card > player_1_card:
        player_2.push(player_1_card)
        player_2.push(player_2_card)
        player = "second"
    count += 1


print(player, count if count <= 1e6 else "botva")

```
