```python
class Node:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

class Stack:
    def __init__(self):
        self.last = None
        self.length = 0

    def push(self, value: int) -> None:
        new_node = Node(value)
        if self.length > 0:
            new_node.prev = self.last
            self.last.next = new_node
        self.last = new_node
        self.length += 1

    def check_stack(self) -> bool:
        return True if self.length else False

    def pop(self) -> None:
        if self.check_stack():
            node = self.last
            if self.length > 1:
                self.last = self.last.prev
                self.last.next = None
                node.prev = None
            self.length -= 1
            return node.value

    def back(self) -> int:
        if self.check_stack():
            return self.last.value

filename = "great_lineland_migration_input_1.txt"

with open(filename, encoding="utf-8") as src:
    n_cities = int(src.readline())              # 2 <= n_cities <= 1e5
    a = list(map(int, src.readline().split()))  # 0 <= a <= 1e9

answer = [-1] * n_cities

stack = Stack()
stack.push((0, a[0]))
for i in range(1, n_cities):
    while stack.back() and stack.back()[1] > a[i]:
        idx, value = stack.pop()
        answer[idx] = i
    stack.push((i, a[i]))

print(*answer)
```
