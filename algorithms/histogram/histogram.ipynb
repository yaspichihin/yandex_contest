```python
filename = 'histogram_input_2.txt'

with open(filename, encoding='utf-8') as src:
    text = ''.join([line.strip().replace(' ', '') for line in src.readlines()])

symbols = {}
max_value = 0
for symbol in text:
    symbols[symbol] = symbols.get(symbol, 0 ) + 1
    if symbols[symbol] > max_value:
        max_value = symbols[symbol]

symbols = dict(sorted(symbols.items(), key=lambda x: ord(x[0])))    

while max_value > 0:
    [print('#', end='') if symbols[i] >= max_value else print(' ', end='')for i in symbols]
    print()
    max_value -= 1
print(''.join(symbols.keys()))
```
