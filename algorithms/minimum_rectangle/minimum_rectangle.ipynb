```python
filename = './minimum_rectangle_input_1.txt'

with open(filename, encoding='utf-8') as src:
    k = int(src.readline().strip())
    points = [tuple(map(int, line.strip().split())) for line in src.readlines()]

points.sort(key=lambda x: x[0])
min_x = points[0][0]
max_x = points[-1][0]

points.sort(key=lambda x: x[1])
min_y = points[0][1]
max_y = points[-1][1]

print(min_x, min_y, max_x, max_y)
```
