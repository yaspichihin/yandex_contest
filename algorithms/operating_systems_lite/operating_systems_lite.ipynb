```python
filename = './operating_systems_lite_input_2.txt'

with open(filename, encoding='utf-8') as src:
    m = int(src.readline())
    n = int(src.readline())
    sectors = [list(map(int, line.split())) for line in src.readlines()]


correct_oss = []
count_oss = 0

for sector in sectors:
    for os in correct_oss.copy():
        if sector[0] <= os[1] and \
            sector[1] >= os[0]:
            correct_oss.remove(os)
            count_oss -= 1
    correct_oss.append(sector)
    count_oss += 1

print(count_oss)
```
