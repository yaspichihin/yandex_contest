```python
class Node:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

class Stack:
    def __init__(self):
        self.last = None
        self.length = 0

    def push(self, value: int) -> None:
        new_node = Node(value)
        if self.length == 0:
            self.last = new_node
            self.length += 1
        else:
            self.last.next = new_node
            new_node.prev = self.last
            self.last = new_node
            self.length += 1

    def check_stack(self) -> bool:
        return True if self.length else False

    def pop(self) -> str:
        if self.check_stack():
            value = self.last.value
            if self.length == 1:
                self.last = None
            else:
                new_last = self.last.prev
                self.last.prev = None
                new_last.next = None
                self.last = new_last
            self.length -= 1
            return value

    def back(self) -> str:
        if self.check_stack():
            return self.last.value

    def size(self) -> int:
        return self.length


stack = Stack()

with open("postfix_notation_input_1.txt", encoding="utf-8") as src:
    seq = [symbol for symbol in src.readline().strip() if symbol != " "]

for symbol in seq:
    if symbol not in "+-*":
        stack.push(symbol)
    else:
        num_2 = stack.pop()
        num_1 = stack.pop()
        stack.push(str(eval(num_1+symbol+num_2)))

print(stack.back())
```
