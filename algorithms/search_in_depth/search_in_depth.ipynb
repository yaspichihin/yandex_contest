```python
# filename = "search_in_depth_input_1.txt"

# with open(filename, encoding="utf-8") as src:
#     n, m = map(int, src.readline().split()) # 1 <= n <= 1e3 and 0 <= m <= 1e5
#     edges = [list(map(int, edge.split())) for edge in src.readlines()]

n, m = list(map(int, input().split()))

connectivity_list = {}
visited_vertexes = {}

for vertex in range(1, n+1):
    connectivity_list[vertex] = []
    visited_vertexes[vertex] = False

for _ in range(m):
    vertex_1, vertex_2 = list(map(int, input().split()))
    connectivity_list[vertex_1] += [vertex_2]
    if vertex_1 != vertex_2:
        connectivity_list[vertex_2] += [vertex_1]


# # Сформируем список связности
# connectivity_list = {}
# # Сформируем посещенные вершины
# visited_vertexes = {}

# for vertex in range(1, n+1):
#     connectivity_list[vertex] = []
#     visited_vertexes[vertex] = False

# for edge in edges:
#     vertex_1, vertex_2 = edge
#     connectivity_list[vertex_1] += [vertex_2]
#     if vertex_1 != vertex_2:
#         connectivity_list[vertex_2] += [vertex_1]


def dfs(graph, visited, now):
    visited[now] = True
    for neighbor in graph[now]:
        if not visited[neighbor]:
            dfs(graph, visited, neighbor)

dfs(connectivity_list, visited_vertexes, 1)

k = 0
k_i = []
for vertex, is_visited in visited_vertexes.items():
    if is_visited:
        k += 1
        k_i.append(vertex)

print(k)
print(*k_i)
```
