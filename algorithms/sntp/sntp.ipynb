```python
filename = './sntp_input_2.txt'

with open(filename, encoding='utf-8') as src:
    a = list(map(int, src.readline().strip().split(':')))
    b = list(map(int, src.readline().strip().split(':')))
    c = list(map(int, src.readline().strip().split(':')))


def get_seconds(time: list[int]):
    return time[2] + time[1] * 60 + time[0] * 3600

a_sec = get_seconds(a)
b_sec = get_seconds(b)
c_sec = get_seconds(c)

# if else если вермя c переходит на следующий день
if a_sec < c_sec:
    correct_time = b_sec + round((c_sec - a_sec) / 2 + 10e-6)
else:
    correct_time = b_sec + round((c_sec + 86400 - a_sec) / 2 + 10e-6)

hours = correct_time // 3600 % 24
minutes = correct_time % 3600 // 60
seconds = correct_time % 60


hours = hours if hours > 9 else '0' + str(hours)
minutes = minutes if minutes > 9 else '0' + str(minutes)
seconds = seconds if seconds > 9 else '0' + str(seconds)

print(f"{hours}:{minutes}:{seconds}")
```
