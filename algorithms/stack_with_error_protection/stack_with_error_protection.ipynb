```python
class Node:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

class Stack:
    def __init__(self):
        self.last = None
        self.length = 0

    def push(self, value: int) -> None:
        """
        Добавить в стек число n (значение n задается после команды).
        Программа должна вывести ok.
        """
        new_node = Node(value)
        if self.length == 0:
            self.last = new_node
            self.length += 1
        else:
            self.last.next = new_node
            new_node.prev = self.last
            self.last = new_node
            self.length += 1
        print("ok")

    def pop(self) -> None:
        """
        Удалить из стека последний элемент.
        Программа должна вывести его значение.
        """
        if self.check_stack():
            value = self.last.value
            if self.length == 1:
                self.last = None
            else:
                new_last = self.last.prev
                self.last.prev = None
                new_last.next = None
                self.last = new_last
            self.length -= 1
            print(value)
        else:
            print("error")

    def back(self) -> None:
        """
        Программа должна вывести значение последнего элемента,
        не удаляя его из стека.
        """
        if self.check_stack():
            print(self.last.value)
        else:
            print("error")

    def size(self) -> None:
        """
        Программа должна вывести количество элементов в стеке.
        """
        print(self.length)

    def clear(self) -> None:
        """
        Программа должна очистить стек и вывести ok.
        """
        self.last = None
        self.length = 0
        print("ok")

    def exit(self) -> None:
        """
        Программа должна вывести bye и завершить работу.
        """
        print("bye")


    def check_stack(self) -> bool:
        """
        Перед исполнением операций back и pop программа должна проверять,
        содержится ли в стеке хотя бы один элемент.
        
        Если во входных данных встречается операция back или pop, и при этом стек пуст,
        то программа должна вместо числового значения вывести строку error.
        """
        return True if self.length else False

stack = Stack()
with open("./stack_with_error_protection_input_4.txt", encoding="utf-8") as src:
    for line in src:
        line = line.strip().split()
        
        command = line[0]
        if len(line) > 1:
            value = int(line[1])
        
        if command == "push":
            stack.push(value)
        elif command == "pop":
            stack.pop()
        elif command == "back":
            stack.back()
        elif command == "size":
            stack.size()
        elif command == "clear":
            stack.clear()
        elif command == "exit":
            stack.exit()
            del stack
            break
```
