```python
filename = "test_input_16.txt"

with open(filename, encoding="utf-8") as src:
    n_students = int(src.readline())    # 2 <= n_students <= 1e9
    n_options = int(src.readline())     # 2 <= n_options <= n_students
    peter_row = int(src.readline())
    peter_choise = int(src.readline())


# n_students      = 10
# n_options       = 6
# peter_row       = 5
# peter_choise    = 2

# Определим вариает Пети
peter_seat= (peter_row - 1) * 2 + peter_choise
peter_option = peter_seat % n_options if peter_seat % n_options else n_options

# Определим есть ли такой вариант перед Петей
is_option_before = True if peter_seat > n_options else False

# Определим есть ли такой вариант после Пети
is_option_after = True if (n_students - peter_seat) // n_options else False


dict_change = {1: 2,
               2: 1}



if not is_option_before and not is_option_after:
    # print(1)
    answer = [-1]

elif not is_option_before and is_option_after:
    # print(2)
    seat_after  = peter_seat + n_options
    row_after = seat_after // 2 + seat_after % 2
    answer = [row_after,  peter_choise if n_options % 2 == 0 else dict_change[peter_choise]]

elif is_option_before and not is_option_after:
    # print(3)
    seat_before = peter_seat - n_options
    row_before = seat_before // 2 + seat_before % 2
    answer = [row_before, peter_choise if n_options % 2 == 0 else dict_change[peter_choise]]

elif is_option_before and is_option_after:
    # print(4)
    seat_before = peter_seat - n_options
    seat_after  = peter_seat + n_options

    row_before = seat_before // 2 + seat_before % 2
    row_after = seat_after // 2 + seat_after % 2

    dist_before = peter_row - row_before
    dist_after = row_after - peter_row

    if dist_after <= dist_before:
        answer = [row_after,  peter_choise if n_options % 2 == 0 else dict_change[peter_choise]]
    else:
        answer = [row_before, peter_choise if n_options % 2 == 0 else dict_change[peter_choise]]

print(*answer)
# 25 тест -> 49926764 2
```
