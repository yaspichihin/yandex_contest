```python
filename = "the_cheapest_way_input_4.txt"

with open(filename, encoding="utf-8") as src:
    n, m = map(int, src.readline().split())
    mtx = [list(map(int, line.split())) for line in src.readlines()]

min_path = [[float("inf") for _ in range(m+1)] for _ in range(n+1)]
min_path[0][1] = 0
min_path[1][0] = 0
for i in range(1, n+1):
    for j in range(1, m+1):
        min_path[i][j] = min(min_path[i-1][j], min_path[i][j-1]) + mtx[i-1][j-1]

print(min_path[i][j])
```
