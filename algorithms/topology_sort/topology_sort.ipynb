```python
filename = "topology_sort_input_10.txt"

with open(filename, encoding="utf-8") as src:
    vs, es = map(int, src.readline().split())
    edges = [list(map(int, row.split())) for row in src.readlines()]

# vs = 7
# edges = [[1, 2], [2, 3], [3, 4], [5, 2], [5, 4], [2, 7]]

conn_list = {}
visited = {}
for vertex in range(1, vs+1):
    conn_list[vertex] = []
    visited[vertex] = (False, 0)

for edge in edges:
    vertex_1, vertex_2 = edge
    conn_list[vertex_1] += [vertex_2]

flag = False
answer = []
def dfs(graph, visited, now):
    visited[now] = (True, 1)
    for neig in graph[now]:
        if visited[neig][0] and visited[neig][1] == 1:
            global flag
            flag = True

        if not visited[neig][0]:
            dfs(graph, visited, neig)
            visited[neig] = (True, 2)
            answer.append(neig)

for vertex, data in visited.items():
    if not data[0]:
        dfs(conn_list, visited, vertex)
        answer.append(vertex)


answer = " ".join(map(str, answer[::-1]))
print(-1 if flag else answer)
```
