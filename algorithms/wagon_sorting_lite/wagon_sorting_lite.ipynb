```python
class Node:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

class Stack:
    def __init__(self):
        self.last = None
        self.length = 0

    def push(self, value: int) -> None:
        node = Node(value)
        if self.length != 0:
            self.last.next = node
            node.prev = self.last
        self.last = node
        self.length += 1

    def check_stack(self) -> bool:
        return True if self.length else False

    def pop(self) -> int:
        if self.check_stack():
            node = self.last
            if self.length == 1:
                self.last = None
            else:
                self.last = self.last.prev
                self.last.next = None
                node.prev = None
            self.length -= 1
            return node.value

    def back(self) -> int:
        if self.check_stack():
            return self.last.value


stack = Stack()
with open("./wagon_sorting_lite_input_3.txt") as src:
    n_wagons = int(src.readline().strip())
    wagons = list(map(int, src.readline().strip().split()))

required_wagon = 1
for wagon in wagons:
    stack.push(wagon)
    while stack.back() == required_wagon:
        stack.pop()
        required_wagon += 1
    
print("NO" if stack.length else "YES")
```
