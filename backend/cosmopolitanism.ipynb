```python
with open('cosmopolitanism_input.txt') as input_file:
    n_countries = int(input_file.readline().strip())
    incomes_required = list(map(int, input_file.readline().strip().split()))
    degrees_required = list(map(int, input_file.readline().strip().split()))
    inheritance_allowed = list(map(int, input_file.readline().strip().split()))

    n_classmates = int(input_file.readline().strip())
    incomes = list(map(int, input_file.readline().strip().split()))
    degrees = list(map(int, input_file.readline().strip().split()))
    inheritances = list(map(int, input_file.readline().strip().split()))


classmate_options = []
for classmate in range(n_classmates):

    countries_options = []
    
    if inheritances[classmate] and inheritance_allowed[inheritances[classmate] - 1]:
        countries_options.append(inheritances[classmate])

    for country_idx in range(n_countries):
        if degrees[classmate]:
            if incomes[classmate] >= incomes_required[country_idx]:
                countries_options.append(country_idx + 1)
                break
        else:
            if not degrees_required[country_idx] and incomes[classmate] >= incomes_required[country_idx]:
                countries_options.append(country_idx + 1)
                break 

    if countries_options:
        classmate_options.append(min(countries_options))
    else:
        classmate_options.append(0)


# Выведем ответ
print(*classmate_options)

# Дальше 12 теста не смог пройти по TL
```
