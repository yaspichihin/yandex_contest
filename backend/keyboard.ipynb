```python
with open('keyboard_input.txt', encoding='utf-8') as input_file:
    # keyboard settings
    buttons = int(input_file.readline().strip())
    symbols = list(map(int, input_file.readline().strip().split()))
    mask = list(map(int, input_file.readline().strip().split()))

    # report settings
    report_length = int(input_file.readline().strip())
    report = list(map(int, input_file.readline().strip().split()))

hash_table = {k: v for k, v in zip(symbols, mask)}

diversity = 0
for i in range(1, report_length):
    if hash_table[report[i]] != hash_table[report[i-1]]:
        diversity += 1

print(diversity)
```
