```python
with open('perfectionism_input_4.txt', encoding='utf-8') as input_file:
    n, x, t = map(int, input_file.readline().strip().split())       # 3 фигуры 5 идеальный вес 2 минуты до праздника
    ices = list(map(int, input_file.readline().strip().split()))    # Идеальный вес фигур 5 10 6

velocity = 1 # Скорость добавления или уменьшения 1 кг льда в мин

# Получим текушую разницу в весе от идеала по модулю, там где успеем еще что-то подправить
diff = [abs(ice - x) for ice in ices]

# Сделаем hash таблицу
hash = {k: v for k, v in enumerate(diff, start=1)}

# Уберем из таблицы значения где точно не успеем привести фигуру в идеал
hash = {k: v for k, v in hash.items() if v <= t}

# Отсортируем таблицу по возростанию требуемого времени
hash = dict(sorted(hash.items(), key = lambda x: x[1]))

# Найдем ответ задачи:
count = 0
figures = []

for figure, required_time in hash.items():
    if t - required_time >= 0:
        t -= required_time * velocity
        count += 1
        figures.append(figure)
    else:
        break

print(count)
print(*sorted(figures))
```
