```python
import numpy as np
import matplotlib.pyplot as plt

with open('purchase_input_1.txt') as input_file:
    n = int(input_file.readline().strip())  # number of days
    prices = list(map(int, input_file.readline().strip().split()))  # prices


# Тестовые данные
# prices1 = [2, 4, 6, 7, 8, 6, 9, 9, 5, 3, 3, 4, 3, 4, 7, 3, 6, 7, 3, 9, 5, 2, 3, 3, 4, 6, 6, 9, 4, 9]
# prices2 = [9, 6, 4, 3, 3, 4, 6, 3, 7, 7, 5, 2, 8, 7, 9, 5, 9, 3, 2, 6, 6, 8, 4, 2, 5, 7, 5, 2, 9, 2]
# prices3 = [6, 1, 8, 4, 7, 3, 8, 7, 1, 9, 9, 2, 4, 4, 8, 6, 8, 3, 6, 6, 7, 4, 4, 7, 5, 2, 4, 5, 7, 3]
# prices4 = [9, 9, 2, 2, 4, 1, 7, 4, 1, 2, 6, 8, 7, 5, 7, 8, 9, 6, 8, 8, 7, 2, 2, 4, 9, 1, 3, 1, 8, 8]
# prices5 = [6, 6, 1, 1 ,6, 5, 7, 3, 4, 9, 6, 2, 6, 1, 7, 9, 4, 9, 6, 9, 8, 3, 3, 8, 4, 6, 9, 8, 3]
# prices6 = [9, 4, 4, 9, 5, 6, 8, 4, 5, 7, 3, 7, 8, 5, 8, 8, 4, 2, 4, 1, 2, 6, 7, 7, 9, 3, 7, 7, 9]
# prices7 = [2, 4, 6, 7, 8, 6, 9, 9, 5, 3, 3, 4, 3, 4, 7, 3, 6, 7, 3, 9, 5, 2, 3, 3, 4, 6, 6, 9, 4, 9]
# prices = prices7


# Получение производной от графика акций
delta_prices = [0] + [prices[i] - prices[i-1]for i in range(1, len(prices))]
delta_len = len(delta_prices)

# Визуализация графика акция и ее производной
# plt.figure(figsize=(15, 10))
# plt.plot(range(1, len(prices) + 1), prices)
# plt.plot(range(1, len(prices) + 1), delta_prices, 'r:')
# plt.hlines(y=0, xmin=1, xmax=len(prices) + 1, color='r')
# plt.xticks(range(1, len(prices)))
# plt.grid()
# plt.show()

# Решение
deal_days = []
stocks = False
for i in range(delta_len - 1):

    # buy - покупать акции
    if not stocks and delta_prices[i] <= 0 and delta_prices[i+1] > 0:
        stocks = True
        deal_days.append(i+1)

    # sell - продавать акции
    if stocks and delta_prices[i] >= 0 and delta_prices[i+1] < 0:
        stocks = False
        deal_days.append(i+1)
        
    # sell в последний день
    if stocks and i == delta_len - 2 and delta_prices[i+1] >= 0:
        stocks = False
        deal_days.append(i+2)

# Вывод решения
print(len(deal_days) // 2)
if len(deal_days) // 2:
    for i in range(len(deal_days) // 2):
        print(*deal_days[i*2:(i+1)*2])
```
