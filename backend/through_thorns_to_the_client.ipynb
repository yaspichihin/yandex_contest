```python
with open('through_thorns_to_the_client_input.txt') as file:
    n = int(file.readline().strip())
    logs = [line.strip().split() for line in file.readlines()]


events_id = {}
for log in logs:
    day, hour, minute, rid = map(int, log[:4])
    event = log[4]

    if event == 'A':
        event_type = 0
    elif event == 'C' or event == 'S':
        event_type = 1
    elif event == 'B':
        continue

    hour = day * 24 + hour
    minute = hour * 60 + minute

    if rid not in events_id:
        events_id[rid] = []
    events_id[rid].append((minute, event_type))

for rid in sorted(events_id):
    events_id[rid].sort()
    ans = 0
    for event in events_id[rid]:
        if event[1] == 0:
            prev_time = event[0]
        elif event[1] == 1:
            ans += event[0] - prev_time 

    print(ans, end=' ')
```
