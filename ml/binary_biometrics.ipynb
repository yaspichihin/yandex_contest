```python
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
from torch.nn import functional as F
from torch.utils.data import TensorDataset, DataLoader
from torchmetrics import Accuracy

import librosa

from tqdm.autonotebook import tqdm
```

```python
# Считывание исходных данных
with open('f:/datasets/yandex_voice_2/binary_train.tsv') as f:
    xs, ys = [], []
    for line in f:
        x, y = line.strip().split()
        xs.append(x)
        ys.append(y)

max_length = 300
for i in range(len(xs)):
    file = np.load(f'f:/datasets/yandex_voice_2/{xs[i]}.npy')[:max_length, :]
    file = file.T
    file = np.pad(file, [[0, 0], [0, max(0, max_length - file.shape[1])]], mode='constant', constant_values='-1')
    xs[i] = torch.from_numpy(file)

ys = torch.from_numpy(np.array(ys, dtype=np.int8))
ys
```

```python
class NN(nn.Module):
    def __init__(self, in_features, out_features):
        super().__init__()

        self.conv1d_1 = nn.Conv1d(n_in = 64, n_out=128, kernel=9, padding=0)
        self.maxpool1d_2  = nn.MaxPool1d (n_in = 128, n_out=128, kernel=2, padding=0)

        self.lin_linear_1 = nn.Linear(in_features, in_features * 4)
        self.lin_linear_2 = nn.Linear(in_features * 4, in_features * 2)
        self.lin_linear_3 = nn.Linear(in_features * 2, out_features)
        self.relu = F.

    def forward(self, x):
        # Указываем в каком порядке работать с слоями
        x = self.lin_linear_1(x)
        x = self.relu(x)            # все значения меньше 0 превратить в 0
        x = self.lin_linear_2(x)
        x = self.relu(x)
        x = self.lin_linear_3(x)
        return x
```
